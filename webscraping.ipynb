```python
import requests
from bs4 import BeautifulSoup

num_pages = 20

base_url = "https://www.amazon.in/s"

params = {
    "k": "bags",
    "crid": "2M096C61O4MLT",
    "qid": "1653308124",
    "sprefix": "ba,aps,283",
    "ref": f"sr_pg_{{page}}"
}

data = []

for page in range(1, num_pages + 1):
    params["ref"] = params["ref"].format(page=page)
    response = requests.get(base_url, params=params)
    soup = BeautifulSoup(response.text, "html.parser")
    product_items = soup.find_all("div", {"data-component-type": "s-search-result"})
    for item in product_items:
        product_url = item.find("a", class_="a-link-normal s-no-outline")["href"]
        product_name = item.find("span", class_="a-size-medium a-color-base a-text-normal").text.strip()
        product_price = item.find("span", class_="a-offscreen").text.strip()
        rating_element = item.find("span", class_="a-icon-alt")
        rating = rating_element.text.strip() if rating_element else ""
        num_reviews_element = item.find("span", {"class": "a-size-base", "dir": "auto"})
        num_reviews = num_reviews_element.text.strip() if num_reviews_element else ""
        data.append({
            "Product URL": product_url,
            "Product Name": product_name,
            "Product Price": product_price,
            "Rating": rating,
            "Number of Reviews": num_reviews
        })
for item in data:
    print(item)

```

```python
import csv
from google.colab import files

output_file = 'scraped_data.csv'
with open(output_file, mode='w', encoding='utf-8', newline='') as file:
    writer = csv.writer(file)
    header_row = ['Product URL', 'Product Name', 'Product Price', 'Rating', 'Number of Reviews']
    writer.writerow(header_row)
    for item in data:
        row = [item['Product URL'], item['Product Name'], item['Product Price'], item['Rating'], item['Number of Reviews']]
        writer.writerow(row)
files.download(output_file)
```

```python
import csv
import requests
from bs4 import BeautifulSoup
from urllib.parse import urlparse, urljoin

file_path = '/content/scraped_data.csv'

# Read the CSV file and extract the URLs
urls = []
with open(file_path, 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        urls.append(row['Product URL'])

# List to store scraped data
scraped_data = []

# Function to scrape product details from a URL
def scrape_product_details(url):
    if not urlparse(url).scheme:
        url = "http://" + url
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    description_element = soup.find('div', {'id': 'productDescription'})
    description_text = description_element.text.strip() if description_element else ''

    asin_element = soup.find('th', string='ASIN')
    asin = asin_element.next_sibling.text.strip() if asin_element else ''

    product_description_element = soup.find('h2', string='Product Description')
    product_description = product_description_element.next_sibling.strip() if product_description_element else ''

    manufacturer_element = soup.find('th', string='Manufacturer')
    manufacturer = manufacturer_element.next_sibling.text.strip() if manufacturer_element else ''

    return description_text, asin, product_description, manufacturer

for url in urls:
    url = urljoin('https://www.amazon.in/', url)
    description, asin, product_description, manufacturer = scrape_product_details(url)
    print("Product URL:", url)
    print("Description:", description)
    print("ASIN:", asin)
    print("Product Description:", product_description)
    print("Manufacturer:", manufacturer)
    print() 

```
